## Python Data types

1. Numeric types

```python
# Integer (int)
age = 36
negative = -34
large = 1_000_000

print(f"Integer: {age}, type: {type(age)}")
print(f"Large_number: {large}")
```

```python
# Float (float)
price = 19.09
pi = 3.1415
scientific = 1.5e3

print(f"Integer: {price}, type: {type(price)}")
```

```python
# Complex
complex_num = complex(2,5)
x = 3 + 6j

print(f"Complex: {x}, type: {type(x)}")
print(f"Real Part: {x.real}, Imaginary part: {x.imag}")
```

## Sequence Types

```python
# String (str)

name = 'python'
multiline = """Virat Kohli[a] (born 5 November 1988) is an Indian international cricketer and the former all-format captain of the Indian national cricket team.[3] He is a right-handed batter and occasional right-arm medium pace bowler. Considered one of the greatest all-format batsmen in the history of cricket, he has been nicknamed the King, the Chase Master, and the Run Machine for his skills, records and ability to lead his team to victory.[4] Kohli has the most centuries in ODIs and the second-most centuries in international cricket with 85 tons across all formats. He is also the leading run-scorer in the Indian Premier League.[5] Kohli is the most successful Test captain of India with most wins and 3 consecutive Test mace retainments.[6] He is the only batter to earn 900+ rating points across all 3 formats.[7]"""

print(f"String: {name}, type: {type(name)}")
print(multiline)
```

```python
# List (list) ordered and mutable
colors = ['blue','red','green']
mixed = [23, 'shashi',True, [9,8],{"name":"shashi"}, ("python"), 89.0]

print(f"List: {colors}, Type: {type(colors)}")
print(mixed)
```

```python
# Tuple (tuple) ordered, immutable

coordinates = (10, 20)
one_num = (34,"hey")
print(f"Tuple : {coordinates}, Type: {type(coordinates)}")
print(one_num)
```

```python
# Range (range) -- sequnece of numbers

number = range(5)
customized_num = range(2,10,2)


print(f"Range: {list(number)}")
print(f"customized_num: {list(customized_num)}")

```

## Mapping Type

```python
# Dictionary (dict) ---key-value pair

employee = {"name":"shahsi","age":36,"city":"Hyderabad"}
nested_one = {"user": {'id':345, 'admin':True}}

print(f"Dictionary: {employee}, Type: {type(employee)}")
print(nested_one)
```

```python
# Set (set) unordered collection of unique numbers

colors = {'blue','red','green'}
numbers = {1,1,3,3,5,7,7,9,10,10}

print(f"Set: {colors}, Type: {type(colors)}")
print("Unique Numbers: {}".format(numbers))
```

## Boolean Type

```python
# Boolean bool - True or False

is_active = True
is_done = False

print(f"Boolean: {is_active}, Type: {type(is_active)}")

print(f"10 > 5: {10 > 5}")
```

```python
# Nonetype -- indicates absence of value

result = None

print(f"None: {result}, Type: {type(result)}")
```
