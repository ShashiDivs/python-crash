## Strings

```python
str_methods = [i for i in dir(str) if not i.startswith('_')]
print(f"Total String methods: {len(str_methods)}\n")
print("All string methods")
for i, method in enumerate(str_methods, 1):
    print(f"{i:2}. {method}")
```

## Case Conversion Methods

```python
text = "Hello python"

# capitalize() first char is upper the rest is lower
print(f"Capitalize(): {text.capitalize()}")

# upper() - All chars are uppercase
print(f"upper(): {text.upper()}")

# lower() - All chars are lowerrcase
print(f"lower(): {text.lower()}")

# title() - First char of each word is upper
print(f"title(): {text.title()}")

# swapcase() - swapping upper to lower and vice versa
print(f"swapcase(): {text.swapcase()}")
```

## Alignment and Padding methods

```python
text = "AgentiAI"

# center() Center align with padding
print(f"center(20, '*'): {text.center(20,'*')}")

#ljust() left align with padding
print(f"ljust(20,'_'): {text.ljust(20, '_')}")

#rjust() left align with padding
print(f"rjust(20,'_'): {text.rjust(20, '_')}")
```

## Searching and Replacing methods

```python
text = "Hey, I am learning GenAI and Agentic AI, I love GenAI"

# find()
print(f"find(GenAI): {text.find('I')}")

#Index() like find the index if not then throws an error
print(f"Index(love): {text.index('GenAI')}")

# count() how many times occurance
print(f"count(G): {text.count('G')}")

# replace
print(f"repplace('GenAI','Python'): {text.replace('GenAI','Python')}")

```

## Splitting and Joining Methods

```python
text = "Hey, I am learning GenAI and Agentic AI, I love GenAI"

# split Split this string into a list
print(f"Split: {text.split()}")

#rsplit() Split fron the right
print(f"rsplit(','): {text.rsplit(',', 1)}")


# join() - Join iterbale with str separator


words = ['Hey','I am', "Learing", "GenAI"]

" ".join(words)
```
