## Lists

- Order - items maintain their position
- Mutable - Can be changed after creation
- Allow Duplicates Same kind of value come many times
- Mixed Types - Can be having multiple data types

```python
list_methods = [i for i in dir(list) if not i.startswith('_')]
print(f"Total list methods: {len(list_methods)}\n")
print("All list methods")
for i, method in enumerate(list_methods, 1):
    print(f"{i:2}. {method}")
```

## Creating list

```python
# Empty list

empty_list = []
print(empty_list)

# list with values
nums = [4,6,8,9,3]
print(nums)

# list with different types
mixed  = [1,'sasi',3.89, True, None]
print(mixed)

ranges = list(range(5))
print(ranges)

dups = [1,2,2,3,3,4,1]
print(dups)

```

## Accesing Elements in the list

```python
colors  = ['red','blue','green','white','purple']

print(colors)

print(f"colors[0]: {colors[0]}")
print(f"colors[3]: {colors[2]}")

print(f"colors[-1]: {colors[-1]}")

print(f"slice colors[1:2]: {colors[1:3]}")
print(f"Reverese: {colors[::-1]}")
```

## Append 

- adding element to the end

```python
# append() example

colors = ['white','red'] # existing one
print(f"before appending: {colors}")

colors.append("blue")
print(f"after appending: {colors}")

colors.append("green")
print(f"after appending: {colors}")


nums = [3,4,5]
nums.append(6)
print(nums)

nums.append([7,8])
print(nums)
```

# Extend

- Adding all elements from iterativbe to the end

```python
# extend() examples

lis1 = [1,2,3]
lis2 = [4,5,6]

print(f"list 1: {lis1}")
print(f"list 2: {lis2}")

lis1.extend(lis2)
print(f"after extended: {lis1}")

lets =['a','b']
lets.append('cd')
print(lets)

lets.extend('cd')
print(lets)
```

## insert()

- inserting element at the specific index

```python
colors = ['green','red','white']
print(colors)

# inserting blue at 1 position
colors.insert(1,'blue')
print(f"After inserted: {colors}")

# inserting purple at 0 position
colors.insert(0,'purple')
print(f"After inserted: {colors}")
```

## remove()

- removes the first occurance of the specific element


```python
colors = ['yellow','green','red','white','blue','yellow']
print(colors)

colors.remove('yellow')
print("After remove: {}".format(colors))
```
