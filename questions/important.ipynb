```python
lst = [1, [2, [3, [4]]], 5,"shashi"]
```

```python
def flatten(lst):

    result = []

    for i in lst:
        if isinstance(i, list):
            result.extend(flatten(i))
        else:
            result.append(i)
    
    return result
```

```python
flatten(lst)
```

```python
dics = { "a": 1, "b": { "d": 3, "e": 5 } }
```

```python
def flatten_dict(d, prefix='', sep='.'):

    result = {}

    for k, v in d.items():

        key = f"{prefix}{sep}{k}" if prefix else k

        print(key)

        if isinstance(v, dict):

            result.update(flatten_dict(v, key, sep))
        else:
            result[key] = v
    
    return result

```

```python
flatten_dict(dics, prefix='', sep='.')
```
